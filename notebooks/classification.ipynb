```python
# main
full_mp = pd.read_parquet("../import/output/mp.parquet")
extras = pd.read_parquet("../import/input/extras.parquet")

class_cols = [
    'rd_no',
    'year_occurred',
    'original_iucr',
    'current_iucr',
    'current_primary_class',
    'current_secondary_class',
    'description',
    'how_located',
    'primary',
    'status',
    'status_date',
    'closed_date',
    'date_officer_arrived',
    'notification_time',
    'date_occurred',
    'beat_occurred',
    'city_beat_occurred',
    'city_district_occurred',
    'city_sector_occurred',]
mp = full_mp[class_cols].copy().sort_values('year_occurred')
```

```python
not_nc = (~mp.primary.str.contains("NON-CRIMINAL", na=False, regex=False)) & \
        (~mp.current_primary_class.str.contains("NON-CRIMINAL", na=False, regex=False))
crim = (mp.primary.str.contains("CRIM", na=False, regex=False)) |  \
       (mp.current_primary_class.str.contains("CRIM", na=False, regex=False))
mp.loc[(not_nc) & (crim), 
       ['rd_no', 'year_occurred', 'status', 'closed_date', 
        'primary', 'current_primary_class', 
        'description', 'current_secondary_class']]
```
